```python
import os

def save_dialogue_to_txt(input_path, output_file):
    with open(output_file, 'w', encoding='utf-8') as output_f:
        for filename in os.listdir(input_path):
            if filename.endswith('.txt'):
                file_path = os.path.join(input_path, filename)
                with open(file_path, 'r', encoding='utf-8') as input_f:
                    lines = input_f.readlines()
                    dialogue = ' '.join([line.strip() for line in lines if not line.strip().startswith("화자") and line.strip()])  # "화자"로 시작하거나 공백인 문장은 건너뜀
                    sentences = split_sentences(dialogue)  # 문장을 '.' 나 '?'를 기준으로 분할
                    output_f.write('\n'.join(sentences) + '\n\n')

def split_sentences(text):
    sentences = []
    start = 0
    for i, char in enumerate(text):
        if char in ['.', '?']:
            sentence = text[start:i+1].strip()
            if sentence:
                sentences.append(sentence)
            start = i+1
    return sentences




input_path = "./voice_phishing_script"
output_file = "dialogue.txt"
save_dialogue_to_txt(input_path, output_file)
```
