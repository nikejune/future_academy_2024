```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
```

```python
import random #너무 빨리 가져가면 차단당할수 있어서1
```

```python
driver = webdriver.Chrome()
```

```python
#그놈목소리
cnt = 0 
for page in range(1, 13) :
    driver.get(f"https://www.fss.or.kr/fss/bbs/B0000203/list.do?menuNo=200686&bbsId=&cl1Cd=&pageIndex={page}")
    time.sleep(random.random() + 1)
    #input() #자동 방지 화면 방지용
    div_list = driver.find_element(By.CSS_SELECTOR, "div.bd-list-thumb-a")
    a_list = div_list.find_elements(By.CSS_SELECTOR, "a")
    
    for a in a_list : 
        a.send_keys(Keys.COMMAND +"\n")  # 맥
       # a.send_keys(Keys.CONTROL +"\n") # 윈도우
        time.sleep(random.random() +  1)
        driver.switch_to.window(driver.window_handles[-1]) 
        time.sleep(random.random() + 1)
        
        file_name = driver.find_element(By.CSS_SELECTOR, "span.file-name")
        file_name.click()
        cnt += 1
        print(driver.current_url, cnt)
        # 현재 사용중인 탭 종료
        driver.close()
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(random.random() + 1)
```

```python
#대출 사기형
cnt = 0 
for page in range(1, 20) :
    driver.get(f"https://www.fss.or.kr/fss/bbs/B0000206/list.do?menuNo=200690&&pageIndex={page}")
    time.sleep(random.random() + 1)
    #input() #자동 방지 화면 방지용
    td_list = driver.find_elements(By.CSS_SELECTOR, "td.title")
    
    for td in td_list : 
        a = td.find_element(By.CSS_SELECTOR, "a")
        a.send_keys(Keys.COMMAND +"\n")  # 맥
       # a.send_keys(Keys.CONTROL +"\n") # 윈도우
        time.sleep(random.random() +  1)
        driver.switch_to.window(driver.window_handles[-1]) 
        time.sleep(random.random() + 1)
        
        video = driver.find_element(By.CSS_SELECTOR, "video")
        video_url = video.get_property('src')
        print(video_url)
        
        driver.get(video_url)
        
        cnt += 1
        
        print(driver.current_url, cnt)
        # 현재 사용중인 탭 종료
        driver.close()
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(random.random() + 1)

```
