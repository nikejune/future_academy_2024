# 빈도 분석(명사)

```python
from konlpy.tag import Kkma
import pandas as pd
```

```python
import pandas as pd

def text_to_dataframe(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append({'Dialogue': line.strip()})

    df = pd.DataFrame(data)
    return df

file_path = "dialogue.txt"
df = text_to_dataframe(file_path)
print(df)
```

```python
kkma = Kkma()
```

```python
total = []
for i in df['Dialogue'] : 
    nouns = kkma.nouns(i)
    total += nouns
total
```

```python
from collections import Counter # 빈도분석

Counter(total).most_common(100) # 가장 많이 나온 단어 TOP 100
```

```python
dic = Counter(total)
dic
```

```python
#1글자 제거
dic = {key: value for key, value in dic.items() if len(key) > 1}

print(dic)
```

```python
# 값(value)에 따라 내림차순으로 정렬하여 상위 20개의 아이템을 추출
dic = dict(sorted(dic.items(), key=lambda x: x[1], reverse=True)[:100])

print(dic)
```

```python
#불용어 제거 지시대명사 등
del dic['지금']
del dic['때문']
del dic['제가']
del dic['이제']
del dic['이것']
del dic['서울']
del dic['일단']
del dic['우리']
del dic['이번']
del dic['이거']
del dic['오늘']
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color="white", 
               font_path = "BMDOHYEON_ttf.ttf",
               colormap = "Blues").generate_from_frequencies(dic)
plt.axis('off')
plt.imshow(wc)
plt.show()
```
